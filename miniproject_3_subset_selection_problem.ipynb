1. Problem Statement:

Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 5

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}


```python
import random as r

Set = set([-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4])
SetSize = 5
ResultList = set()
Iterations = 10000

for i in range(Iterations):
    Chromosome = r.sample(Set, SetSize)
    Chromosome.sort()
    if sum(Chromosome) == 0:
        ResultList.add(tuple(Chromosome))

for r in ResultList:
    print(r)

print("\nTotal Sets:", len(ResultList))
```

2. Problem Statement:

Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 3 to 6 only

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
from itertools import combinations

# Given set
S = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
S_list = list(S)

# Result list to store valid subsets
result = []

# Check subsets of size 3 to 6
for size in range(3, 7):
    for subset in combinations(S_list, size):
        if sum(subset) == 0:
            result.append(subset)

# Print valid subsets
for subset in result:
    print(subset)

# Print total number of valid subsets
print("\nTotal subsets with sum = 0:", len(result))
```
